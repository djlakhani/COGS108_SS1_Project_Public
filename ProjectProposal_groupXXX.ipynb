# Research Question


Our research question aims to determine if hosting the Olympics truly beneficial for a city or country. Although we have not finalized our question, we will choose from the potential questions below based on the data we are able to obtain:

- What is the scope of the economic impact of the Olympics across countries over time? Are these changes significant? Are these changes sustained? To what extent does economic impact affect societal quality (employment rate, literacy rate, etc.) of life within the host cities?

# Background and Prior Work


The Olympics is an athletic competition hosted by a country every four years that brings spectators and travelers from all over the world to see their favorite countries competing. To look good to the international community, countries will often invest millions in hosting the event and the surrounding communities. After the pressure is off of the hosting country, it is unsure whether a country economically benefits from the event or stagnates due to debt.
There are a couple articles and scientific journals published that talk about this topic. One study found that “regional per capita GDP significantly increases by 3.6 %-points (3.3 %-points) relative to national per capita GDP in the year of the event (the year before the event)” [^Firgo2021]. On the other hand, another article said that the games “provide little in return” [^Vassalo2021] and affect the impoverished the most, but there doesn’t seem to be references to data to back up these statements.

In line with the varying perspectives on the economic impact of hosting the Olympics, further research provides further insight into the subject. While hosting the Olympics can lead to short-term economic growth through increased tourism and infrastructure development, it also comes with substantial costs and uncertain long-term benefits. For instance, the 2016 Rio Olympics generated an estimated $1.8 billion in tourism-related spending, showcasing the potential economic benefits [^Wills2021]. However, the 2012 London Olympics had a total budget of approximately $14.6 billion, highlighting the significant financial investment required [^Wills2021].
The need for careful planning, financial management, and a clear strategy for post-Olympic use of facilities is essential to ensure that the long-term economic benefits outweigh the costs associated with hosting such a massive event. These statistics demonstrate the complex nature of the economic impact of hosting the Olympics, indicating the importance of a comprehensive analysis considering both short-term gains and long-term sustainability. All in all, there are many different opinions on this topic and there are many different, unexplored avenues to answer this question, such as through a human/economic lens.

In our final project, we will make the extra effort to pick datasets from reputable sources in order to ensure the utmost accuracy in our data.
